# Array operations with NumPy

In this notebook we'll explore how NumPy arrays make math operations fast and easy:

- Element-wise operations
- Broadcasting
- Comparison to Python Lists

### Element wise operations

```python
import numpy as np
```

```python
a_arr = np.array([1, 2, 3, 4])
b_arr = np.array([10, 20, 30, 40])

print(a_arr)
print(b_arr)
```

```python
# element wise operations
print(a_arr + b_arr)
print(a_arr - b_arr)
print(a_arr * b_arr)
print(a_arr / b_arr)
print(a_arr ** 2) # square each element in a_arr
print(np.sqrt(a_arr)) # square root using sqrt method
```

### Broadcasting

Broadcasting allows operations between arrays of different shapes.

NumPy automatically "expands" arrays when possible.

```python
my_arr = np.array([1, 2, 3, 4, 5])

print("arr + 10:", my_arr + 10)
```

```python
print("arr multiplied by 2:", my_arr * 2)
```

```python
mat = np.array([[1, 2, 3],[4, 5, 6]])
print(mat + np.array([10, 20, 30]))
```

### Comparing NumPy arrays to Python lists

```python
import time

size = 10_000_000

# Python list
L = list(range(size))
start = time.time()
L2 = [x * 2 for x in L]
print("Python list time:", time.time() - start)

# NumPy array
A = np.arange(size)
start = time.time()
A2 = A * 2
print("NumPy array time:", time.time() - start)
```
