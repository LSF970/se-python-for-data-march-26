# Intro to the Pandas library

```python
import numpy as np
import pandas as pd
```

- Pandas is a Python library for data manipulation and analysis
- Built on top of NumPy
- Series = 1D data structure --> Like a column
- DataFrame = 2D data structure --> Like a spreadsheet, SQL table etc.
- Always explore data first before doing analysis (shape, columns, rows, dtypes, stats)

[Pandas documentation](https://pandas.pydata.org/docs/user_guide/index.html#user-guide)

### Series

```python
# creating a series
data = [10, 20, 30, 40, 50]
series = pd.Series(data)
print(series)
```

```python
# accessing values and index
print(series.values)
print(series.index)
```

```python
temps = pd.Series([12, 18, 22, 19, 25])

print("Max temp:", temps.max())
print("Min temp:", temps.min())
```

Notes:
- Series behave like a list but has an index
- .values() gives raw data, index gives the labels

### Dataframes

```python
# creating a dataframe
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['London', 'Paris', 'Berlin', 'Madrid']
}

df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
data1 = {
   "Product": ["Book", "Pen", "Notebook"],
   "Price": [10, 2, 5],
   "Quantity": [3, 10, 7]
}
df1 = pd.DataFrame(data1)
print(df1) # first part

# Add a new column for total value
df1["Total Value"] = df1["Price"] * df1["Quantity"]
print(df1)
```

### Reading data from a csv file

```python
# load a dataset from a local csv
students_df = pd.read_csv("students_with_scores.csv")

# print df first 5 rows
print(students_df.head())
```

```python
# print df last 5 rows
print(students_df.tail())

# print last row of df
print(students_df.tail(1))
```

### Exploring data with Pandas

```python
# Check the df shape
print(students_df.shape)

# Column names
print(students_df.columns)

# Summary statistics
print(students_df.describe())

# Info about dtypes and missing values in df
print(students_df.info())
```

### loc() and iloc()

- loc() uses names to get the locations of values --> so strings bascially
- iloc() does the same but using numbers --> Uses index positions only
- Can mix and match the two

```python
# Single column
print(students_df["Name"])
```

```python
# Multiple columns
print(students_df[["Name", "School"]])
```

```python
# Row by label - using the index labels not index position
print(students_df.loc[0])
```

```python
print(students_df.loc[0:2])
```

```python
print(students_df.loc[0:2, ["Name", "Grade"]])
```

```python
# iloc()
print(students_df.iloc[0])
```

```python
print(students_df.iloc[0:3])
```

Examples of loc() vs iloc()

```python
# All students from a specific school
print(students_df.loc[students_df["School"] == "West High"])
```

```python
# Scores of the last 2 students to take the assessment
print(students_df.iloc[-2:, students_df.columns.get_loc("Score")])
```

### Useful functions for filtering

```python
# Sort --> ascendeing by default
print(students_df.sort_values(by="Score"))
```

```python
# Sort --> descending by setting ascending to False
print(students_df.sort_values(by="Score", ascending=False))
```

```python
# Filter by numeric value
print(students_df[students_df["Score"] >= 80])
```

```python
# Filter by numeric value, sort ascending
print(students_df[students_df["Score"] >= 80].sort_values(by="Score", ascending=False))
```

```python
# Average score by school
print(students_df.groupby("School")["Score"].mean())
```
