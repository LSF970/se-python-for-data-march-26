# Data Cleaning with Pandas

```python
import numpy as np
import pandas as pd
```

```python
# import messy data
df = pd.read_csv("student_data_messy.csv")
print(df.head())
```

### Step 1: Dealing with missing values

- Option 1 = Drop rows/columns that contain missing values
- Option 2 = Fill missing values with some kind of value (static value, avg etc.)

### Option 1 - dropping missing values

```python
# Drop ANY row that has at least 1 missing value
df_dropped = df.dropna()
print(df_dropped.head())
```

```python
# Drop rows  where a specific column has a missing value
df_dropped_col = df.dropna(subset=["Age"])
print(df_dropped_col.head(10))
```

### Option 2 - Filling missing values

```python
# Fill missing age value with a static number - in this case make new column for "Age_Filled"
df["Age_Filled"] = df["Age"].fillna(0)
df
```

```python
# Filling a missing value with an average - mean or median etc.
df["Score_Filled"] = df["Score"].fillna(df["Score"].mean())
df.head(20)
```

### Step 2 - Filtering and sorting

- Filterning helps explore subsets of data that meet specific conditions
- Sorting helps quickly find highest/lowest values

```python
# Make a new df with only those over the age of 17
over_17_df = df[df["Age"] > 17]
over_17_df.head(5)
```

```python
# Filter for multiple conditions
high_score = df[(df["Age"] > 17) & (df["Score"] > 90)]
high_score.head()
```

```python
# Sorting by multiple columns
df_sorted_multi = df.sort_values(by=["Age", "Score"], ascending=[True, False])
df_sorted_multi.head(15)
```

### Step 3 - Renaming columns

```python
# rename a specific column
df_renamed = df.rename(columns={"Age": "Age_Years", "Score": "Test_Score"})
df_renamed.head()
```
